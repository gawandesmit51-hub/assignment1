#3. Grade Calculator:
#Write a program to input marks and assign grades as:

```python
marks=float(input("Enter your marks:"))
if marks<0 or marks>100
    print("Marks are not eligible ")
elif marks>=90:
    print("YOU GOT 'A' GRADE")
elif marks>=80:
    print("YOU GOT 'B' GRADE")
elif marks>=70:
    print("YOU GOT 'C'GRADE")



```
