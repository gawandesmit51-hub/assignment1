#Discount Calculator:
#Input the price of an item and apply discounts:

```python
amount=int(input("Enter your amount"))
if amount>100:
    discount=(amount/100)*20
    print(f"YOU WILL GET ${discount}(20%) discount  on ${amount}")
elif 50<=amount<=100:
    discount=(amount/100)*10
    print(f"YOU WILL GET ${discount}(10%) discount on ${amount}")
else:
    print(" Sorry no discoun for below $50 shopping")
```
